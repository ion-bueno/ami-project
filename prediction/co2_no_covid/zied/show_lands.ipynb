```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import pearsonr

path = "/Users/ziedbk/Desktop/group01/datasets"
df_CO2 = pd.read_csv(path+'/greenhouse/co2_country_edgar.csv')

years = np.linspace(1970, 2018, 49)
country1 = df_CO2.loc[df_CO2['country_name'] == 'Germany']
country2 = df_CO2.loc[df_CO2['country_name'] == 'Australia']

country1 = country1.iloc[0][1:].str.replace(',', '.', regex=True).astype(float)
country2 = country2.iloc[0][1:].str.replace(',', '.', regex=True).astype(float)
plt.figure(figsize=(27, 12))
plt.plot(years, country2)
plt.plot(years, country1)
plt.legend(['Germany','Australia'], fontsize= 24)
p = sns.jointplot(country1, country2,);
pearsonr(country1, country2)[0]
```
